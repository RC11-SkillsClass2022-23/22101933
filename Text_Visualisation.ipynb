```python
combined_paragraphs = []
paragraph_book_index = []

# split the book into saperate paragraphs
paragraphs = file_content.split('\n\n')
paragraphs_2 = [x for x in paragraphs if len(x)>100] # only pick short paragraphs

for p in paragraphs_2:
  combined_paragraphs.append(p)
```

```python
stoplist = set('for a of the end to in and'.split(' ')) # get rid of words with no actual meaning and split the sentences
texts = [[word.replace('.','').replace(',','') for word in document.lower().split() if (word not in stoplist) and len(word)>2] for document in combined_paragraphs] # split the sentence to words (also filter out meaningless words)

to_delete = []
for i in range(len(texts)):
  t = texts[i]
  test = [word for word in t if word.isalpha()]
  if len(test)<20:
    to_delete.append(i)
  else:
    texts[i]=test

for i in sorted(to_delete,reverse = True):  # delete the short sentence
  del texts[i]
  del combined_paragraphs[i]

```

```python
from collections import defaultdict
frequency = defaultdict(int)
for t in texts:
  for token in t:
    frequency[token] += 1

processed_corpus = [[token for token in t if frequency [token]>1] for t in texts]# get rid of all words in paragraphs that only happen once.

# I get rid of all of the words in the paragraphs that only happen once, because then they're not really relevant that because then, if something only occurs once, and there's no way of actually relating anything else to it. So that's why I then
```

In the code above, a defaultdict named frequency is created to store the token frequencies. Then, for each text in texts, the inner loop iterates over each token in the text, and the frequency of the token is incremented in the frequency dictionary.

The processed corpus is created using a list comprehension that iterates over each text in texts. For each text, it filters the tokens based on their frequency using the condition if frequency[token] > 1, and stores the resulting tokens in a sublist.

The resulting processed_corpus will be a list of sublists, where each sublist contains tokens that have a frequency greater than 1

```python
bow_corpus = [dictionary.doc2bow(text) for text in processed_corpus] #j processed all sentences
```

```python
# This is a TF-IDF model using the TfidfModel class from the gensim models, it computes the TF-IDF scores for each term in the corpus
from gensim import models

tfidf = models.TfidfModel(bow_corpus)
```

```python
words = 'Standing on the roof of Istanbul'.lower().split() # This line converts the input sentence to lowercase and splits it into a list of individual words. The resulting list will contain the words:
print(tfidf[dictionary.doc2bow(words)])
# Each tuple represents a word in the input sentence along with its TF-IDF score. The word ID corresponds to the position of the word in the dictionary, and the TF-IDF score represents the importance or relevance of the word in the context of the entire corpus.
```

```python
index_tmpfile = get_tmpfile('index')
index = Similarity(index_tmpfile, bow_corpus, num_features = len(dictionary))
```

```python
# input the sentence you want to find its similar one
query_document = 'The Grand Bazaar, one of the largest and oldest markets in the world'.lower().split()
query_bow = dictionary.doc2bow(query_document)

sims = index[query_bow]
# The sims variable will contain a list of similarity scores. The position of each score in the list corresponds to the index of the document in the bow_corpus. You can use these scores to rank the documents based on their similarity to the query document.
```

```python
for document_number, score in sorted(enumerate(sims), key = lambda x:x[1], reverse = True):
  print(document_number, score)
```
